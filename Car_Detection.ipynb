```python
from ultralytics import YOLO

  # Load model (start from YOLOv8n pretrained weights)
model = YOLO('yolov8n.pt')  # you can change to yolov8s.pt, yolov8m.pt if you want bigger model later

    # Train model
model.train(
        data='data.yaml',
        imgsz=640,
        epochs=50,
        batch=16,
        patience=30,
        optimizer='AdamW',
        lr0=0.01,
        lrf=0.1,
        device=0,
        degrees=10,
        translate=0.1,
        scale=0.1,
        shear=0.1,
        perspective=0.0005,
        flipud=0.0,
        fliplr=0.5,
        hsv_h=0.015,
        hsv_s=0.1,
        hsv_v=0.1,
        mosaic=0.5,
        mixup=0.1,
        project='car_model_yolov8',
        name='exp',
        exist_ok=True)
```

```python
from ultralytics import YOLO
model = YOLO("/content/drive/MyDrive/car_detection/car_model_yolov8/exp/weights/last.pt")
model.train(resume=True)
```
